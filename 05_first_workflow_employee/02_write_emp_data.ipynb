```python
dbutils.widgets.text('dept_name', '')
```

```python
_dept = dbutils.widgets.get('dept_name')
```

```python
df = spark.read.format("csv") \
    .option("header", "true") \
    .load("/databricks-datasets/retail-org/company_employees/company_employees.csv")

df.show()
```

```python
from pyspark.sql.functions import * 
from pyspark.sql.types import * 

df_filtered = df.filter(upper(col('department')) == _dept.upper())
df_filtered.show()
```

```python
_count = df_filtered.count()
if _count >0:
    df_filtered.write.mode('overwrite').saveAsTable(f'dev.bronze.dept_{_dept}')
    print(f"data written for {_dept} department")
    display(spark.table(f"dev.bronze.dept_{_dept}"))
else:
    print(f'no data for {_dept}')
```
