```python
data = [("East", "Jan", 200), ("East", "Feb", 300), 
        ("East", "Mar", 250), ("West", "Jan", 400), 
        ("West", "Feb", 350), ("West", "Mar", 450)]

columns = ["Region", "Month", "Sales"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

window_spec = Window.partitionBy('Region').orderBy('Sales')
window_spec_rank = Window.partitionBy('Region').orderBy(desc('Sales'))

df_cumm = df.withColumn('cumm_total',sum('Sales').over(window_spec))
df_rank = df.withColumn('rank_num',rank().over(window_spec_rank))
df_cumm.show()
df_rank.show()
```
