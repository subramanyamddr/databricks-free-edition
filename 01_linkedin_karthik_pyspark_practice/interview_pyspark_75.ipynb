## Level1 - `Ingestion`

```python
from pyspark.sql import Row

data = [
    Row(User_ID=101, User_Activity='{"device":"Android","app_version":"v5.2","city":"Delhi","viewed_items":["Shoes","Watch"]}'),
    Row(User_ID=102, User_Activity='{"device":"iOS","app_version":"v5.3","city":"Mumbai","viewed_items":["T-shirt","Bag","Shoes"]}'),
    Row(User_ID=103, User_Activity='{"device":"Android","app_version":"v5.1","city":"Bangalore","viewed_items":["Laptop"]}'),
    Row(User_ID=104, User_Activity='{"device":"iOS","app_version":"v5.2","city":"Chennai","viewed_items":["Phone","Earphones"]}')
]

df = spark.createDataFrame(data)
display(df)
```

### Level 2 (nested)

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
from pyspark.sql.window import Window

schema = StructType([
    StructField('device',StringType(),True),
    StructField('app_version',StringType(),True),
    StructField('city',StringType(),True),
    StructField('viewed_items',ArrayType(StringType()),True)
])

df_res = df.withColumn('json',from_json("User_Activity",schema))
df_final = df_res.select("User_ID",col("json.device"),col("json.app_version"),col("json.city"),col("json.viewed_items").alias('viewed_items'))\
    .withColumn('items',explode('viewed_items'))
display(df_final)
```
