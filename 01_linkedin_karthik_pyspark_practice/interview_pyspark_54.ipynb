```python
data = [
    "Hello world",
    "Hello from PySpark",
    "PySpark is awesome",
    "Hello PySpark world"
]

df = spark.createDataFrame(data, "string").toDF("line")
display(df)

words_df = df.selectExpr("explode(split(line, ' ')) as word")
word_count_df = words_df.groupBy("word").count()
display(word_count_df)
```
