```python
data = [
    (101, 1001, "2025-01-15", 500.00),
    (102, 1002, "2025-01-14", 300.00),
    (101, 1001, "2025-01-17", 550.00),
    (103, 1003, "2025-01-16", 450.00),
    (102, 1002, "2025-01-18", 320.00),
    (103, 1003, "2025-01-19", 460.00)
]

schema = ["customer_id", "order_id", "order_date", "total_amount"]

df = spark.createDataFrame(data, schema)
display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

window_spec = Window.partitionBy('new_id').orderBy(desc('order_date'))

df_comp = df.withColumn('new_id',concat("customer_id",lit("_"),"order_id")).withColumn(
    "rank_num", row_number().over(window_spec)
).filter(col("rank_num") == 1).drop("new_id","rank_num")
df_comp.show()
```
