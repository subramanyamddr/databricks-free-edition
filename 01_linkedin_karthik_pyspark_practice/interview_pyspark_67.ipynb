```python
data = [
    ("C001", "2024-06-01", 1000),
    ("C001", "2024-06-05", 1500),
    ("C001", "2024-06-07", 2000),
    ("C002", "2024-06-02", 5000),
    ("C002", "2024-06-03", 7000),
    ("C002", "2024-06-08", 6500),
    ("C003", "2024-06-01", 3000),
    ("C003", "2024-06-04", 4000)
]

columns = ["customer_id", "txn_date", "amount"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

window_spec = Window.partitionBy('customer_id').orderBy(desc('txn_date'))
df_rank = df.withColumn('rank',row_number().over(window_spec)).where(col('rank') <= 2)\
            .withColumn('lag',lag('amount').over(window_spec))\
            .withColumn('diff',col('lag')-col("amount"))
df_rank.show()
```
