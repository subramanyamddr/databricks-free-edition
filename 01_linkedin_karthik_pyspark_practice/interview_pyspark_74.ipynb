```python
data = [
    (101, "Mumbai", 4.8, 1200),
    (102, "Delhi", 4.2, 900),
    (103, "Chennai", 3.2, 700),
    (104, "Bangalore", None, 500),
    (105, "Kolkata", 4.6, 1500)
]

columns = ["Restaurant_ID", "City", "Avg_Rating", "Orders"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
from pyspark.sql.window import Window

df_res = df.withColumn('label',
                       when((col('Avg_Rating')>=4.5) & (col('Orders')>1000),"Excellent")\
                      .when((col('Avg_Rating').between(3.5,4.5)),"Good")\
                      .when((col('Avg_Rating')<3.5),"Needs Improvement")\
                          .otherwise('Not Rated')    
                       )
df_res.show()
```
