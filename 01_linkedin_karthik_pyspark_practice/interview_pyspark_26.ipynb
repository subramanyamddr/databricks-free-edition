```python
data = [
    (1, "apple,banana,orange"),
    (2, "mango,grapes"),
    (3, "pineapple")
]
columns = ["ID", "Tags"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.functions import *

df_exp = df.withColumn('tag',explode(split("Tags",","))).select('ID','tag')
df_exp.show()
```
