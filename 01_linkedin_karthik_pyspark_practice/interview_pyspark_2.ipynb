```python
data = [ ("Store A", "2024-01", 800), ("Store A", "2024-02", 1200), ("Store A", "2024-03", 900), ("Store B", "2024-01", 1500), ("Store B", "2024-02", 1600), ("Store B", "2024-03", 1400), ("Store C", "2024-01", 700), ("Store C", "2024-02", 1000), ("Store C", "2024-03", 800) ] 

df = spark.createDataFrame(data, ["Store", "Month", "Sales"]) 

df.show()
```

```python
df.dtypes
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import sum
window_spec = Window.partitionBy('Store').orderBy('Month')

df_cumm_sales = df.withColumn('cumm_sales',sum('Sales').over(window_spec))
df_cumm_sales.show()

df_total_sales = df.groupBy('Store').agg(sum('Sales').alias('total_sales'))
df_total_sales.show()
```

```python
df_final_sales = df_cumm_sales.join(df_total_sales,on="Store").select('Store','total_sales','cumm_sales')
df_final_sales.distinct().show()
# df_final_sales.select('Store','total_sales','cumm_sales').show()
```
