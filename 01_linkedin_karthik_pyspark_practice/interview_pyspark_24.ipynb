```python
data = [
    {
        "id": 1,
        "name": "Alice",
        "department": "HR",
        "address": {"city": "New York", "state": "NY"}
    },
    {
        "id": 2,
        "name": "Bob",
        "department": "IT",
        "address": {"city": "San Francisco", "state": "CA"}
    },
    {
        "id": 3,
        "name": "Charlie",
        "department": "Finance",
        "address": {"city": "Chicago", "state": "IL"}
    }
]

df = spark.createDataFrame(data)
display(df)
```

```python
df.dtypes
```

```python
df_expel = df.select('id', 'name', 'department', df['address']['city'].alias('city'), df['address']['state'].alias('state'))
df_expel.show()
```
