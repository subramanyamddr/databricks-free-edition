```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

data = [
    ("C001", "2024-06-01 10:00"),
    ("C001", "2024-06-01 10:05"),
    ("C001", "2024-06-01 10:30"),
    ("C002", "2024-06-01 09:00"),
    ("C002", "2024-06-01 09:12"),
    ("C002", "2024-06-01 09:13")
]
columns = ["customer_id", "login_time"]

df = spark.createDataFrame(data, columns).withColumn('login_time', to_timestamp('login_time'))
display(df)
```

```python
df.dtypes
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

window_spec = Window.partitionBy('customer_id').orderBy('login_time')

df_dates = df.withColumn('lag_date',lag('login_time').over(window_spec))\
             .withColumn('date_diff',(unix_timestamp('login_time')-unix_timestamp('lag_date'))/60)\
             .where(col('date_diff')<10)
display(df_dates)
```
