```python
data = [
    (1, "Laptop", 1000, 5),
    (2, "Mouse", None, None),
    (3, "Keyboard", 50, 2),
    (4, "Monitor", 200, None),
    (5, None, 500, None),
]

columns = ["product_id", "product", "price", "quantity"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.functions import *

mean_val = df.select(mean('price')).collect()[0][0]
mean_val

df_work = df.na.fill({'price':mean_val,'quantity':1}).dropna(subset=['product'])
df_work.show()
```
