```python
sales_data = [
    (101, "P001", 100),
    (101, "P002", 200),
    (101, "P003", 150),
    (102, "P004", 300),
    (103, "P005", 400),
    (101, "P006", 500),
    (104, "P007", 250)
]

sales_df = spark.createDataFrame(sales_data, ["store_id", "product_id", "amount"])

# Sample Data (Small Dimension Table)
store_data = [
    (101, "Walmart"),
    (102, "Target"),
    (103, "Costco"),
    (104, "BestBuy")
]

store_df = spark.createDataFrame(store_data, ["store_id", "store_name"])

display(sales_df)
display(store_df)
```

```python
spark.conf.get('spark.sql.shuffle.partitions')
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window
import random

@udf
def fn_getRange():
    return random.randint(0, 3)


salt_df = spark.range(0,4)
display(salt_df)
```

```python
df_stored_df = store_df.join(salt_df,how="cross").withColumn('new_store_id',concat('store_id',lit('_'),"id"))
display(df_stored_df)
```

```python
sales_df_df = sales_df.withColumn('new_store_id',concat("store_id",lit("_"),fn_getRange()))
sales_df_df.show()
```
