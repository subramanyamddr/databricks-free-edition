```python
data = [
    (101, "North", 1500),
    (102, "South", 2200),
    (103, "East", 800),
    (104, "North", 1200),
    (105, "West", 900),
    (106, "South", 3000),
    (107, "East", 500),
    (108, "West", 1800),
    (109, "South", 2500),
]

columns = ["customer_id", "region", "purchase_amount"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python

from pyspark.sql.functions import *

df_res = df.groupBy("region").agg(
    count(lit(1)).alias('total_customers'),
    sum("purchase_amount").alias('total_purchase_amount'),
    avg("purchase_amount").alias('avg_purchase_amount'),
    max("purchase_amount").alias('max_purchase'),
    min("purchase_amount").alias('min_purchase')
)

display(df_res)
```
