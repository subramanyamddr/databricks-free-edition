```python
data = [
    ("HR", "2020", 2500), 
    ("HR", "2021", 3200), 
    ("HR", "2022", 2800), 
    ("Engineering", "2020", 5000), 
    ("Engineering", "2021", 6000), 
    ("Engineering", "2022", 5500), 
    ("Marketing", "2020", 4000), 
    ("Marketing", "2021", 3500), 
    ("Marketing", "2022", 3300)
]

# Create DataFrame
df = spark.createDataFrame(data, ["Department", "Year", "Salary"])
display(df)
```

```python
from pyspark.sql.functions import *
df_group = df.groupBy('Department').agg(avg('Salary')).filter(col('avg(Salary)') > 3000)
display(df_group)
```

```python
from pyspark.sql.functions import *
df_total = df.groupBy('Department').agg(sum('Salary').alias('total_sal')).orderBy(desc('total_sal'))
display(df_total)
```
