```python
data = [
    (101, "Aakash", "2024-12-01", 1500),
    (101, "Aakash", "2024-12-10", 2000),
    (102, "Bhavana", "2024-11-15", 500),
    (102, "Bhavana", "2024-12-01", 1200),
    (103, "Charan", "2024-12-05", 2500)
]

columns = ["user_id", "user_name", "txn_date", "amount"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

salary_spec = Window.partitionBy('user_name').orderBy(desc('txn_date'))
df_res = df.withColumn('row_num',row_number().over(salary_spec)).where(col('row_num')==1)
display(df_res)
```
