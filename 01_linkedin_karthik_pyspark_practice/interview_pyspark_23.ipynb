```python
data = [
    (1, "John Doe", "IT", 80000, "2021-01-15"),
    (2, "Jane Smith", "HR", 70000, "2020-05-22"),
    (3, "Robert Brown", "IT", 85000, "2019-11-03"),
    (4, "Emily Davis", "Finance", 90000, "2022-07-19"),
    (5, "Michael Johnson", "HR", 75000, "2023-03-11"),
]

columns = ["EmployeeID", "Name", "Department", "Salary", "JoinDate"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.functions import *
df_empCount = df.groupBy('Department').agg(count("EmployeeID").alias('emp_count'),avg('Salary').alias('avg_salary'))
df_empCount.show()
```

```python
df_result = df.join(df_empCount,on="Department",how="left_outer").filter((col('JoinDate')>lit('2020-01-01')) & (col('Department')==lit("HR")))
df_result.show()
```
