```python
data = [
    (1, 101, 500.0, "2024-01-01"), 
    (2, 102, 200.0, "2024-01-02"), 
    (3, 101, 300.0, "2024-01-03"), 
    (4, 103, 100.0, "2024-01-04"), 
    (5, 102, 400.0, "2024-01-05"), 
    (6, 103, 600.0, "2024-01-06"), 
    (7, 101, 200.0, "2024-01-07"),
]
columns = ["transaction_id", "user_id", "transaction_amount", "transaction_date"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
df.dtypes
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

df_dates = df.withColumn('transaction_date',to_date('transaction_date','yyyy-MM-dd'))
```

```python
window_spec = Window.partitionBy('user_id').orderBy('transaction_date')
df_fdates = df_dates.withColumn('prev_date',lag('transaction_date').over(window_spec))
df_furdates = df_fdates.withColumn('date_diff',date_diff('transaction_date','prev_date'))
df_furdates.show()
```

```python
df_avg = df_furdates.groupBy('user_id').agg(avg('date_diff').alias('avg_col'))
df_avg.show()
```
