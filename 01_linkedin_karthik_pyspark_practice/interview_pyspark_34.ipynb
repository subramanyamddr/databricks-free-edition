```python
data = [("S1", "P1", 100), ("S1", "P2", 200), 
        ("S1", "P3", 300), ("S2", "P1", 400), 
        ("S2", "P2", 100), ("S2", "P3", 500)]

columns = ["StoreID", "Product", "Sales"]

df = spark.createDataFrame(data, columns)

display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

df_pv = df.groupBy('StoreID').pivot('Product').max('Sales')
df_pv.show()
```

```python
df_total = df_pv.withColumn('total_sales',(col("P1")+col("P2")+col("P3")))\
                .withColumn('p1contribution',(col("P1")/col('total_sales')*100))\
                .withColumn('p2contribution',(col("P2")/col('total_sales')*100))\
                .withColumn('p3contribution',(col("P3")/col('total_sales')*100))
df_total.show()
```
