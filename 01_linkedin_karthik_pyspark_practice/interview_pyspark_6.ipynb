```python
data = [ (1, "Laptop", "2024-08-01"), (1, "Mouse", "2024-08-05"), (2, "Keyboard", "2024-08-02"), (2, "Monitor", "2024-08-03") ]
columns = ["customer_id", "product", "purchase_date"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

window_spec = Window.partitionBy('customer_id').orderBy(col('purchase_date').desc())

df_order = df.withColumn('row_num',row_number().over(window_spec)).filter(col('row_num')==1)
display(df_order)
```
