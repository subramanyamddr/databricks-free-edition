```python
customers = spark.read.table("samples.bakehouse.sales_customers")

# Estimate the number of rows needed for ~100MB
sample_size_bytes = customers.limit(1000).toPandas().memory_usage(deep=True).sum()
rows_per_mb = 1000 / (sample_size_bytes / (1024 * 1024))
target_rows = int(rows_per_mb * 100)

customers_100mb = customers.limit(target_rows)

display(customers_100mb)
```

```python
import sys

orders_size_bytes = sys.getsizeof(customers_100mb.toPandas())
orders_size_mb = orders_size_bytes / (1024 * 1024)
orders_size_gb = orders_size_bytes / (1024 * 1024 * 1024)

print(f"Orders DataFrame size: {orders_size_mb:.2f} MB ({orders_size_gb:.4f} GB)")
```
