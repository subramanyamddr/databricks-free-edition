```python
data = [
 ("U001", "2024-01-01"),
 ("U001", "2024-01-02"), # Day 1
 ("U001", "2024-01-08"), # Day 7
 ("U001", "2024-01-31"), # Day 30
 ("U002", "2024-01-01"),
 ("U002", "2024-01-03"),
 ("U003", "2024-01-02"),
 ("U003", "2024-01-03"), # Day 1
 ("U003", "2024-02-01"), # Day 30
]

# Create DataFrame
df = spark.createDataFrame(data, ["user_id", "login_date"])
```

```python
display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

window_spec = Window.partitionBy('user_id').orderBy('login_date')

df_dates = df.withColumn('min_date',min('login_date').over(window_spec)).withColumn('date_diff',date_diff('login_date','min_date'))
df_dates.show()
```

```python
df_result = df_dates.where(col("date_diff").isin([1,7,30]))
df_result.show()
```
