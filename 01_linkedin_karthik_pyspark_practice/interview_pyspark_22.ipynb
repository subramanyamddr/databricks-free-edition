```python
data = [
    ("North", "Laptop", 2000, "Q1"), ("South", "Laptop", 3000, "Q1"), ("East", "Laptop", 2500, "Q1"),
    ("North", "Phone", 1500, "Q1"), ("South", "Phone", 1000, "Q1"), ("East", "Phone", 2000, "Q1"),
    ("North", "Laptop", 3000, "Q2"), ("South", "Laptop", 4000, "Q2"), ("East", "Laptop", 3500, "Q2"),
    ("North", "Phone", 2500, "Q2"), ("South", "Phone", 1500, "Q2"), ("East", "Phone", 3000, "Q2"),
]

columns = ["Region", "Product", "Sales", "Quarter"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
df.dtypes
```

```python
df_pivot = df.groupBy('Product').pivot('Region').sum('Sales')
df_pivot.show()
```

```python
from pyspark.sql.functions import col, round

df_sales = df_pivot\
    .withColumn('Total', (col('East') + col('North') + col('South')))\
    .withColumn('%North', round(col('North')/col('Total')*100, 2))\
    .withColumn('%East', round(col('East')/col('Total')*100, 2))\
    .withColumn('%South', round(col('South')/col('Total')*100, 2))
df_sales.show()
```
