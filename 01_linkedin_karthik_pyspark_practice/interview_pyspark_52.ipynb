```python
data = [("East", "Jan", 200), ("East", "Feb", 300), 
        ("East", "Mar", 250), ("West", "Jan", 400), 
        ("West", "Feb", 350), ("West", "Mar", 450)]

columns = ["Region", "Month", "Sales"]

df = spark.createDataFrame(data, columns)

display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

window_spec = Window.partitionBy('Region').orderBy('Month')
rank_spec = Window.partitionBy('Region').orderBy(desc('Sales'))


df_runn = df.withColumn('cummu_sales',sum('Sales').over(window_spec)).withColumn('rank_sales',rank().over(rank_spec))
display(df_runn)
```
