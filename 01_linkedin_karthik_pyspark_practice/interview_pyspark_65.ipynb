```python
data = [
    ("ABCDE1234F", "SBI", "2024-06-01 10:00:00", 9000),
    ("ABCDE1234F", "ICICI", "2024-06-15 11:30:00", 12000),
    ("ABCDE1234F", "HDFC", "2024-06-20 09:45:00", 15000),
    ("ABCDE1234F", "SBI", "2024-06-25 08:15:00", 11000),
    ("ABCDE1234F", "BOI", "2024-06-27 14:05:00", 5000),
    ("XYZAB9999Z", "SBI", "2024-06-05 13:30:00", 7000),
    ("XYZAB9999Z", "ICICI", "2024-06-12 15:30:00", 8000),
    ("XYZAB9999Z", "HDFC", "2024-06-20 17:30:00", 9500),
    ("XYZAB9999Z", "SBI", "2024-06-29 12:30:00", 6500),
]

columns = ["pan", "bank", "txn_time", "amount"]
df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

window_spec = Window.partitionBy('pan').orderBy(desc('amount'))
df_result = df.withColumn('row_num',row_number().over(window_spec)).where(col('row_num').isin(1,2,3))
df_result.show()

```
