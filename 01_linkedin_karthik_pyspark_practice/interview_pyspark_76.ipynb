```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, ArrayType

data = [
    (101, "Rahul", ["Pizza", "Burger", "Coke"]),
    (102, "Priya", ["Biryani", "Gulab Jamun"]),
    (103, "Aman", ["Sandwich"]),
    (104, "Neha", ["Pasta", "Coffee", "Cake"])
]

schema = StructType([
    StructField("Order_ID", IntegerType(), True),
    StructField("Customer_Name", StringType(), True),
    StructField("Food_Items", ArrayType(StringType()), True)
])

df = spark.createDataFrame(data, schema)
display(df)
```

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
from pyspark.sql.window import Window

df_exp = df.withColumn('food_explode',explode('Food_Items'))
df_exp.show()
```

posexplode we can't use with withColumn, we're getting error. because withColumn expects one column, whereas posexplode returns 2 columns i.e., pos and col.


```python
df_pos_exp = df.select(col("Order_ID"), col("Customer_Name"), posexplode("Food_Items").alias("food_pos", "food_val"))
df_pos_exp.show()
```
