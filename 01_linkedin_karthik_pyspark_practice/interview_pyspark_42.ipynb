```python
data = [
    (1, "Alice", None),
    (2, "Bob", 1),
    (3, "Charlie", 1),
    (4, "David", 2),
    (5, "Eva", 2),
    (6, "Frank", 3),
    (7, "Grace", 3)
]

columns = ["employee_id", "employee_name", "manager_id"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window
import random

df_self = df.alias('A').join(df.alias('B'),col("A.manager_id")==col("B.employee_id"),"left").select('A.employee_id',"A.employee_name",col("B.employee_name").alias('Manager_name'))
df_self.show()
```
