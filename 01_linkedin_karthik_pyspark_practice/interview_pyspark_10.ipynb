```python
data = [("A", "p1", 100), ("A", "p2", 200), ("A", "p3", 200), ("B", "p4", 300), ("B", "p5", 150), ("B", "p6", 150), ("C", "p7", 400), ("C", "p8", 300), ("C", "p9", 200)]
columns = ["category", "product", "amount"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

sales_spec = Window.partitionBy('category').orderBy(desc(col('amount')),col('product'))

df_rank = df.withColumn('dense_rank',row_number().over(sales_spec))

df_final = df_rank.filter(col('dense_rank') <=3)
display(df_final)

```
