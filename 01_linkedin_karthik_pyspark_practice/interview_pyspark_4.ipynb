```python
data = [
    (123, "A", 30, 31, 32),
    (124, "B", 40, 41, 42)
]

columns = ["StudentID", "StudentName", "AScore", "BScore", "CScore"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
unpivot_df = df.selectExpr("StudentID", "StudentName", "stack(3, 'AScore', AScore, 'BScore', BScore, 'CScore', CScore) as (Subject, Score)")
unpivot_df.show()
```
