```python
data1 = [
    (101, "Alice", 500, "2024-08-01"),
    (102, "Bob", 300, "2024-08-02"),
    (103, "Charlie", 450, "2024-08-03")
]
columns1 = ["customer_id", "name", "amount_spent", "transaction_date"]
df1 = spark.createDataFrame(data1, columns1)

data2 = [
    (101, "Alice", 520, "2024-08-01"),
    (104, "David", 600, "2024-08-04"),
    (105, "Eva", 700, "2024-08-05")
]
columns2 = ["customer_id", "name", "total_spent", "transaction_date"]
df2 = spark.createDataFrame(data2, columns2)

display(df1)
display(df2)
```

```python
from pyspark.sql.functions import *

x = ['df1_'+x for x in df1.columns]
y = ['df2_'+y for y in df2.columns]

df1_new = df1.toDF(*x)
df2_new = df2.toDF(*y)
display(df1_new)
display(df2_new)
```

```python
df_join = df1_new.join(df2_new, df1_new['df1_customer_id'] == df2_new['df2_customer_id'], how="outer")

df_final = df_join.na.fill({'df1_customer_id':0,'df1_name':'not applicable','df1_amount_spent':0,'df1_transaction_date':'not applicable','df2_customer_id':0,'df2_name':'not applicable','df2_total_spent':0,'df2_transaction_date':'not applicable'})
display(df_final)
```
