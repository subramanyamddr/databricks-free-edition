```python
data = [
    (101, "Rajesh Kumar", "2023-01-01"),
    (101, "Rajesh Kumar", "2023-05-12"),
    (102, "Anjali Mehta", "2023-02-15"),
    (103, "Rakesh Yadav", "2023-01-20"),
    (103, "Rakesh Yadav", "2023-06-30")
]

columns = ["emp_id", "emp_name", "update_date"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

salary_spec = Window.partitionBy('emp_id').orderBy(desc('update_date'))
df_res = df.withColumn('row_num',row_number().over(salary_spec)).where(col('row_num')==1)
display(df_res)
```
