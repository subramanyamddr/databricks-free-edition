```python
from pyspark.sql import functions as F

# Sample Indian login data
data = [
    ("U001", "2024-01-01"),
    ("U001", "2024-01-02"),
    ("U001", "2024-01-05"),
    ("U001", "2024-01-06"),
    ("U002", "2024-01-03"),
    ("U002", "2024-01-07")
]

# Create DataFrame
df = spark.createDataFrame(data, ["user_id", "login_date"]) \
          .withColumn("login_date", F.to_date("login_date"))

df.show()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

window_spec = Window.partitionBy('user_id').orderBy('login_date')
df=df.withColumn('prev_dates',lag('login_date').over(window_spec))\
    .withColumn('gap_days',date_diff('login_date','prev_dates'))\
        .where(col("gap_days")>1)
display(df)
```
