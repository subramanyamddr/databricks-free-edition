```python
data = [("2023-12-31",), ("31-12-2023",), 
        ("2023/12/31",), ("2024-01-01",), 
        ("13-01-2023",), ("invalid",), ("2022-02-28",)]

columns = ["input_date"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
df.dtypes
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

# Use try_to_date to avoid errors for invalid formats
# Only handle 'yyyy-MM-dd' format as in original logic

df_dates = df.withColumn('new_date',
    when(
        try_to_date(col("input_date"), 'yyyy-MM-dd').isNotNull(),
        date_format(try_to_date(col("input_date"), "yyyy-MM-dd"), "MM/dd/yyyy")
    ).otherwise('Invalid Date')
)
df_dates.show()
```
