```python
data = [
    (1, "Ravi", "IT", 85000),
    (2, "Priya", "HR", 65000),
    (3, "Suresh", "IT", 92000),
    (4, "Neha", "Finance", 78000),
    (5, "Anil", "HR", 72000),
    (6, "Divya", "Finance", 88000),
    (7, "Kiran", "IT", 88000)
]
columns = ["emp_id", "emp_name", "department", "salary"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

salary_spec = Window.partitionBy('department').orderBy(desc('salary'))
#rank_spec = Window.partitionBy('Region').orderBy(desc('Sales'))
df_rank = df.withColumn("row_num",row_number().over(salary_spec)).where(col('row_num')==1).orderBy("department")
display(df_rank)
```

```python
#another simple method

df_res = df.groupBy("department").agg(max("salary").alias("max_sal"))
display(df_res)
```
