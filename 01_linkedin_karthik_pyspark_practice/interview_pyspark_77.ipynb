```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Initialize Spark Session
spark = SparkSession.builder \
    .appName("ZomatoColumnStandardization") \
    .getOrCreate()

# Sample Data
data = [
    (101, "Bengaluru", 450, "2025-01-05 14:21"),
    (102, "Mumbai", 300, "2025-01-05 14:29"),
    (103, "Delhi", 700, "2025-01-05 14:35")
]

columns = ["cust_id", "city", "amount", "order_timestamp"]

df = spark.createDataFrame(data, columns)

print("Original DataFrame:")
df.show()

# ----------------------------------------------------
# 1. Renaming columns using withColumnRenamed()
# ----------------------------------------------------
df1 = df.withColumnRenamed("cust_id", "Customer_ID") \
        .withColumnRenamed("amount", "Order_Amount") \
        .withColumnRenamed("order_timestamp", "Order_Timestamp")

print("After Renaming:")
df1.show()

# ----------------------------------------------------
# 2. Renaming using select() + alias()
# ----------------------------------------------------
df2 = df.select(
    col("cust_id").alias("Customer_ID"),
    col("city").alias("City"),
    col("amount").alias("Order_Amount"),
    col("order_timestamp").alias("Order_Timestamp")
)

print("Using alias():")
df2.show()

# ----------------------------------------------------
# 3. Reordering columns
# ----------------------------------------------------
final_order = ["Customer_ID", "City", "Order_Amount", "Order_Timestamp"]

df_final = df2.select(final_order)

print("Final Reordered DataFrame:")
df_final.show()
```
