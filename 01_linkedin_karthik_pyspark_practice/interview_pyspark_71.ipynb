```python
data = [
    (1, 25, 'North', 'M', '2025-01-01', 150),
    (2, None, 'East', None, '2025-01-02', None),
    (3, 30, 'South', 'F', None, 200),
    (4, 22, None, 'M', '2025-01-03', 180),
    (5, 28, 'West', 'F', None, None)
]

columns = ['Customer_ID', 'Age', 'Region', 'Gender', 'Last_Visit', 'Purchase_Amount']

df = spark.createDataFrame(data, columns)
display(df)
```

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *

num_cols = [x.name for x in df.schema.fields if isinstance(x.dataType ,NumericType)]
str_cols = [x.name for x in df.schema.fields if isinstance(x.dataType ,StringType)]
```

```python
str_cols
```

```python
num_cols
```

```python
for x in num_cols:
    median_val = df.approxQuantile(x,[0.5],0.0)[0]
    df=df.withColumn(x,when(col(x).isNull(),median_val).otherwise(col(x)))
df.show()
```
