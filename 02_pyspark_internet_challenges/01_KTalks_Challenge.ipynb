```python
data = [ 
    (1, [{"amount": 45}, {"amount": 60}]), 
    (2, [{"amount": 30}, {"amount": 20}]), 
    (3, [{"amount": 120}, {"amount": 80}]) 
]
df = spark.createDataFrame(data, ["user_id", "transactions"])
display(df)
```

```python
df.printSchema
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

df_exp = df.withColumn('new_col',explode('transactions')).select('user_id','new_col.amount')
df_res = df_exp.groupBy('user_id').agg(sum("amount").alias('total_sum'))
df_res.show()
```

```python
from pyspark.sql.functions import explode, col, sum

df_exp = df.withColumn("transaction", explode("transactions"))
df_sum = df_exp.groupBy("user_id").agg(sum(col("transaction.amount")).alias("total_sum"))
display(df_sum)
```
